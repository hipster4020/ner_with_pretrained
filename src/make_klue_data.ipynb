### klue dataset

```python
def klue_dataloader():
    with codecs.open("/path/data//klue-ner-v1.1_train.tsv", "r", "utf-8-sig") as io:
        lines = io.readlines()
        result = []
        temp_tokens = []
        temp_tags = []
        temp_string = ''
        temp_tag = ''
        state = True
        for line in lines:
            if line[0] == '#':
                continue
            temp_txt = line.rstrip().split('\t')
            if len(temp_txt) > 1:
                token, tag = temp_txt
                if state:
                    temp_tag = tag
                    temp_string += token
                    state = False
                    continue
                if not re.findall(r'[^a-zA-Z가-힣ㄱ-ㅎ0-9]', token) and token != ' ':
                    if temp_tag[-2:] == tag[-2:]:
                        temp_string += token
                    else:
                        if temp_string:
                            temp_tokens.append(temp_string)
                            temp_tags.append(temp_tag[-2:])
                            temp_string = ''
                            temp_string += token
                            temp_tag = tag
                        else:
                            temp_string += token
                            temp_tag = tag
                            pass
                else:
                    if temp_string:
                        temp_tokens.append(temp_string)
                        temp_tags.append(temp_tag[-2:])
                        temp_string = ''
                        temp_tag = 'O'
            else:
                temp_tokens.append(temp_string)
                temp_tags.append(temp_tag[-2:])
                result.append([temp_tokens, temp_tags])
                temp_tokens, temp_tags = [], []
                temp_string = ''
                temp_tag = ''
    return pd.DataFrame(result, columns=['tokens', 'ner_tags'])
```

# *naver data*

```python
from collections import Counter, defaultdict
```

```python
def ner_tag_processing(li):
    li = list(map(lambda x: re.sub(r"(-[BI])|([BI]-)|(_[BI])|([BI]_)", "", x), li))
    li = list(map(lambda x: "_" if len(x) == 1 and (x == "O" or x == "-") else x, li))
    return li
```

```python
def naver_corpus_data_loader():
    with codecs.open("/path/train_data.txt", "r", "utf-8-sig") as io:
        lines = io.readlines()
        temp = defaultdict(list)
        result = []
        for line in lines:
            if line[0] == "\n":
                result.append(temp)
                temp = defaultdict(list)
                continue
            else:
                temp_line = line.split()
                temp["tokens"].append(temp_line[1])
                temp["ner_tags"].append(temp_line[2])
    naver_df = pd.DataFrame(result)
    naver_df["ner_tags"] = naver_df["ner_tags"].apply(ner_tag_processing)
    return naver_df
```

```python
naver = naver_corpus_data_loader()
```

```python
naver.head()
```

### klue 양식으로 변경

```python
tag_dict = {'LC': 'LOC',
            "DT": "DAT",
            "OG": "ORG",
            "TI": "TIM",
            "QT": "NOH",
            'O': '_',
            "PS": "PER"}

keys = list(tag_dict.keys())
values = list(tag_dict.values())
```

```python
def klue_replace(x):
    x2 = []
    for k, v in enumerate(x):
        if v in values:
#             x2[k] = keys[values.index(v)]
            x2.append(keys[values.index(v)])
        else:
            # x2[k] = 'O' 
            x2.append('O')
    return x2
```

```python
naver['ner_tagss'] = naver.ner_tags.apply(klue_replace)
```

```python
import swifter
from pshmodule.processing import processing as p
```

```python
def processing(x):
    result = []
    for i in x:
        i = re.sub(r"[a-zA-Z가-힣]+.kr", "", i)
        i = re.sub(r"[-=+,#/\?:^.@*\"※~ㆍ!』‘|\(\)\[\]`\'…》\”\“\’·◇▶－()（）．]", "", i)
        result.append(i)
    return result
```
